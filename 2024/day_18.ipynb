```python
# Ingestion


def leggi_file(nome_file):

    out = []
    try:
        with open(nome_file, 'r') as file:
            for riga in file:
                out.append(list(map(int, riga.split(','))))
                
    except FileNotFoundError:
        print(f"Errore: Il file {nome_file} non Ã¨ stato trovato.")
    except ValueError as e:
        print(f"Errore durante la conversione dei dati: {e}")
    return out


nome_file = 'input_day18.txt'
nome_file_test= 'input_day18_test.txt'
data= leggi_file(nome_file)
data_test=leggi_file(nome_file_test)
```

```python
# Funzioni parte 1

import networkx as nx


def crea_grafo(lista, size):
    G = nx.Graph()

    for i in range(size):
        for j in range(size):
            if [i,j] not in lista:
                G.add_node((i,j))
    for i in range(size):
        for j in range(size):
            if (i,j) in G.nodes() and  (i+1,j) in G.nodes():
                G.add_edge((i,j), (i+1,j))
            if (i,j) in G.nodes() and  (i,j+1) in G.nodes():
                G.add_edge((i,j), (i,j+1))
    return G   
```

```python
# Calcolo parte 1

answer_1=0
mappa=data
size=71
caduti=1024

grafo=crea_grafo(mappa[:caduti], size)

answer_1=len(nx.dijkstra_path(grafo, (0,0), (70,70)))-1
print(answer_1)
```
