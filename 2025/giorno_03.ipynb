```python
#Import data


def leggi_file_txt(nome_file):

    lista_righe = []
    try:
        with open(nome_file, 'r') as file:
            for riga in file:
                lista_righe.append(list(map(int, list(riga[:-1]))))
    except FileNotFoundError:
        print(f"Errore: Il file {nome_file} non Ã¨ stato trovato.")
    except ValueError as e:
        print(f"Errore durante la conversione dei dati: {e}")
    return lista_righe

nome_file = 'input_day3.txt'
nome_file_test= 'input_day3_test.txt'
data= leggi_file_txt(nome_file)
data_test=leggi_file_txt(nome_file_test)
```

```python
#Funzioni parte 1

def max_joltage(lista_dati):
    joltage=0
    tens=0
    temp=0
    for i in range(len(lista_dati)-1):
        if lista_dati[i]>tens:
            tens=lista_dati[i]
            temp=i
    joltage=tens*10+max(lista_dati[temp+1:])
    
    return joltage

    
```

```python
#Parte 1

out_1=0

for row in data:

    out_1+=max_joltage(row)
    
print(out_1)
```

```python
#Funzioni parte 2

def from_list_to_int(lista):
    out=0
    for i in range(len(lista)):
        out+=lista[i]*(10**(len(lista)-i-1))
        
    return out
        


def max_joltage_2(lista_dati):
    joltage=[]
    temp=-1
    for k in range(12):
        massimo=0
        for i in range(temp+1, len(lista_dati)-11+k):
            if lista_dati[i]>massimo:
                massimo=lista_dati[i]
                temp=i
        joltage.append(massimo)
        
    return from_list_to_int(joltage)
```

```python
#Parte 2 

out_2=0

for row in data:
    out_2+=max_joltage_2(row)
    
print(out_2)
```
