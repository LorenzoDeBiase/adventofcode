```python
# Ingestion


def leggi_file_txt(nome_file):

    out=[]
    try:
        with open(nome_file, 'r') as file:
            for riga in file:
             out.append([item for item in  riga[:-1].split(' ') if len(item)>0])


                
    except FileNotFoundError:
        print(f"Errore: Il file {nome_file} non è stato trovato.")
    except ValueError as e:
        print(f"Errore durante la conversione dei dati: {e}")
    

    return out


nome_file = 'input_day6.txt'
nome_file_test= 'input_day6_test.txt'
data= leggi_file_txt(nome_file)
data_test=leggi_file_txt(nome_file_test)
```

```python
#Funzioni parte 1

def trasposta(matrice):
    out=[]
    for i in range(len(matrice[0])):
        nuova_riga=[]
        for j in range(len(matrice)):
            nuova_riga.append(matrice[j][i])
        out.append(nuova_riga)
    return out
```

```python
#Parte 1

out_1=0
input=data
for riga in trasposta(input):
    tmp=int(riga[0])
    if riga[-1]=='+':
        for i in range(1,len(riga)-1):
            tmp+=int(riga[i])
    elif riga[-1]=='*':
        for i in range(1,len(riga)-1):
            tmp*=int(riga[i])
    out_1+=tmp

print(out_1)
```

```python
# Ingestion 2


def leggi_file_txt_2(nome_file):

    out=[]
    try:
        with open(nome_file, 'r') as file:
            for riga in file:
             out.append(list(riga[:-1]))


                
    except FileNotFoundError:
        print(f"Errore: Il file {nome_file} non è stato trovato.")
    except ValueError as e:
        print(f"Errore durante la conversione dei dati: {e}")
    

    return out


nome_file = 'input_day6.txt'
nome_file_test= 'input_day6_test.txt'
data= leggi_file_txt_2(nome_file)
data_test=leggi_file_txt_2(nome_file_test)
```

```python
#Funzioni parte 2

def list_to_str(lista):
    out=''
    for item in lista:
        out+=item
    return out.replace(' ', '')
```

```python
#Parte 2

out_2=0
input=trasposta(data)
operazioni=[item[-1] for item in input if item[-1] in ['*', '+']]
numeri=[]

tmp=[]
for item in input:
    if list_to_str(item[:-1])!='':
        tmp.append(int(list_to_str(item[:-1])))
    else:
        numeri.append(tmp)
        tmp=[]
numeri.append(tmp)   

for i in range(len(operazioni)):
    app=numeri[i][0]
    if operazioni[i]=='+':

        for j in range(1,len(numeri[i])):
            app+=numeri[i][j]
    elif operazioni[i]=='*':
        for j in range(1,len(numeri[i])):
            app*=numeri[i][j]
    out_2+=app

print(out_2)

```
