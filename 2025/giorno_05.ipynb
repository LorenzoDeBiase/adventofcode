```python
# Ingestion


def leggi_file_txt(nome_file):

    lista_intervalli = []
    lista_ingredienti=[]
    try:
        with open(nome_file, 'r') as file:
            for riga in file:
                lista_intervalli.append(list(riga[:-1].split('-')))
        for i in range(len(lista_intervalli)):
            if len(lista_intervalli[i])<2:
                temp=i
                break
        lista_ingredienti=[int(item[0]) for item in lista_intervalli[i+1:]]
        lista_intervalli=[tuple(map(int, item)) for item in lista_intervalli[:i]]


                
    except FileNotFoundError:
        print(f"Errore: Il file {nome_file} non Ã¨ stato trovato.")
    except ValueError as e:
        print(f"Errore durante la conversione dei dati: {e}")
    

    return [lista_intervalli,lista_ingredienti]


nome_file = 'input_day5.txt'
nome_file_test= 'input_day5_test.txt'
data= leggi_file_txt(nome_file)
data_test=leggi_file_txt(nome_file_test)
```
