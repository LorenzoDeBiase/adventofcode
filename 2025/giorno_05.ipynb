```python
# Ingestion


def leggi_file_txt(nome_file):

    lista_intervalli = []
    lista_ingredienti=[]
    try:
        with open(nome_file, 'r') as file:
            for riga in file:
                lista_intervalli.append(list(riga[:-1].split('-')))
        for i in range(len(lista_intervalli)):
            if len(lista_intervalli[i])<2:
                temp=i
                break
        lista_ingredienti=[int(item[0]) for item in lista_intervalli[i+1:]]
        lista_intervalli=[tuple(map(int, item)) for item in lista_intervalli[:i]]


                
    except FileNotFoundError:
        print(f"Errore: Il file {nome_file} non Ã¨ stato trovato.")
    except ValueError as e:
        print(f"Errore durante la conversione dei dati: {e}")
    

    return [lista_intervalli,lista_ingredienti]


nome_file = 'input_day5.txt'
nome_file_test= 'input_day5_test.txt'
data= leggi_file_txt(nome_file)
data_test=leggi_file_txt(nome_file_test)
```

```python
#Funzioni parte 1

def in_interval(numero, intervallo):
    if numero>=intervallo[0] and numero<=intervallo[1]:
        return True
    else:
        return False
```

```python
#Parte 1

out_1=0
intervalli, ingredienti = data[0], data[1]

for ingrediente in ingredienti:
    fresh=False
    for intervallo in intervalli:
        if in_interval(ingrediente, intervallo):
            fresh=True
            break
    if fresh==True:
        out_1+=1
print(out_1)
```

```python
# Funzioni parte 2
def conta_elementi(intervallo):
    return intervallo[1]-intervallo[0]+1

def interseca(x,y):
    if x[0]<=y[0]:
        a,b =x , y
    else:
        a, b =y,x
    
    if a[0]< b[0] and a[1]<b[0]:
        return False
    else:
        return True

```

```python
# Parte 2

out_2=0

intervalli=data[0]

lista_unione=[]
for intervallo in intervalli:

    intersecati=[]
    for item in lista_unione:
        if interseca(intervallo, item):
            intersecati.append(item)
    for item in intersecati:
        lista_unione.remove(item)

    
    if intersecati==[]:
        lista_unione.append(intervallo)
    else:       
        if intervallo[0]<=intersecati[0][0]:
            a_1=intervallo[0]
        else:
            a_1=intersecati[0][0]
            
        if intervallo[1]>=intersecati[-1][1]:
            a_2=intervallo[1]
        else:
            a_2=intersecati[-1][1]
        lista_unione.append((a_1,a_2))
    
    lista_unione.sort(key= lambda x: x[0])

for intervallo in lista_unione:
    out_2+=conta_elementi(intervallo)

print(out_2)
            
    
```
