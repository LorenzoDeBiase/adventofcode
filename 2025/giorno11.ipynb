```python
# Ingestion


def leggi_file_txt(nome_file):

    lista_righe = []
    try:
        with open(nome_file, 'r') as file:
            for riga in file:
                lista_righe.append(riga[:-1].replace(':','').split(' '))
                
    except FileNotFoundError:
        print(f"Errore: Il file {nome_file} non Ã¨ stato trovato.")
    except ValueError as e:
        print(f"Errore durante la conversione dei dati: {e}")
    return  lista_righe


nome_file = 'input_day11.txt'
nome_file_test= 'input_day11_test.txt'
data= leggi_file_txt(nome_file)
data_test=leggi_file_txt(nome_file_test)
```

```python
#Parte 1

import networkx as nx 

out_1=0
input=data

G=nx.DiGraph()
for riga in input:
    G.add_node(riga[0])

for riga in input:
    for i in range(1,len(riga)):
        G.add_edge(riga[0],riga[i])

out_1=nx.all_simple_paths(G,'you','out')
print(out_1)
```

```python
#Funzioni parte 2

import collections


def n_percorsi(grafo, inizio, fine):
    generazioni=nx.topological_generations(grafo)
    contatore = collections.Counter()
    contatore[inizio] = 1  

    for generazione in generazioni:
        for individuo in generazione:
            for successore in G.successors(individuo):
                    contatore[successore] += contatore[individuo]
    
    return contatore[fine]
```

```python
#Parte 2

import networkx as nx 

out_2=0
input=data

G=nx.DiGraph()
for riga in input:
    G.add_node(riga[0])

for riga in input:
    for i in range(1,len(riga)):
        G.add_edge(riga[0],riga[i])

srv_fft=n_percorsi(G,'svr', 'fft')
fft_dac=n_percorsi(G, 'fft', 'dac')
dac_out=n_percorsi(G,'dac', 'out')


out_2=srv_fft*fft_dac*dac_out

print(out_2)
```
