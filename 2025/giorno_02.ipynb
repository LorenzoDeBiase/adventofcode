```python
# Ingestion


def leggi_file_txt(nome_file):

    lista_righe = []
    try:
        with open(nome_file, 'r') as file:
            for riga in file:
                lista_app=riga[:-1].split(',')
            for riga in lista_app:
                lista_righe.append(riga.split('-'))
                
    except FileNotFoundError:
        print(f"Errore: Il file {nome_file} non Ã¨ stato trovato.")
    except ValueError as e:
        print(f"Errore durante la conversione dei dati: {e}")
    return  lista_righe


nome_file = 'input_day2.txt'
nome_file_test= 'input_day2_test.txt'
data= leggi_file_txt(nome_file)
data_test=leggi_file_txt(nome_file_test)
```

```python
#Funzioni parte 1


def is_invalid(stringa):
    half=int(len(stringa)/2)
    if stringa[:half]==stringa[half:]:
        return True
    else:
        return False
    
```

```python
#Parte 1

out_1=0
for riga in data:
    for id in range(int(riga[0]), int(riga[1])+1):
        if is_invalid(str(id)):
            out_1+=id
print(out_1)
```

```python
#Funzioni parte 2

def is_invalid_2(stringa):
    half=int(len(stringa)/2)+1
    for i in range(1,half):
        k=0
        sotto_stringhe=[]
        while(k<len(stringa)):
           sotto_stringhe.append(stringa[k:k+i])
           k+=i
        if len(set(sotto_stringhe))==1:
            return True
    else:
        return False
```

```python
#Parte 2

out_2=0
for riga in data:
    for id in range(int(riga[0]), int(riga[1])+1):
        if is_invalid_2(str(id)):
            out_2+=id
print(out_2)
```
