```python
#Import data


def leggi_file_txt(nome_file):

    lista_righe = []
    try:
        with open(nome_file, 'r') as file:
            for riga in file:
                lista_righe.append(list(map(int, riga[:-1].split(','))))
    except FileNotFoundError:
        print(f"Errore: Il file {nome_file} non Ã¨ stato trovato.")
    except ValueError as e:
        print(f"Errore durante la conversione dei dati: {e}")
    return lista_righe

nome_file = 'input_day9.txt'
nome_file_test= 'input_day9_test.txt'
data= leggi_file_txt(nome_file)
data_test=leggi_file_txt(nome_file_test)
```

```python
#Funzioni parte 1

def area(X,Y):
    out=(abs(X[0]-Y[0])+1)*(abs(X[1]-Y[1])+1)
    return out
```
