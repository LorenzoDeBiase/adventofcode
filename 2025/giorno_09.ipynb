```python
# Ingestion


def leggi_file_txt(nome_file):

    lista_righe = []
    try:
        with open(nome_file, 'r') as file:
            for riga in file:
                lista_righe.append(list(map(int, riga[:-1].split(','))))
                
    except FileNotFoundError:
        print(f"Errore: Il file {nome_file} non Ã¨ stato trovato.")
    except ValueError as e:
        print(f"Errore durante la conversione dei dati: {e}")
    return  lista_righe


nome_file = 'input_day9.txt'
nome_file_test= 'input_day9_test.txt'
data= leggi_file_txt(nome_file)
data_test=leggi_file_txt(nome_file_test)
```

```python
#Funzioni parte 1

def area(X,Y):
    out=(abs(X[0]-Y[0])+1)*(abs(X[1]-Y[1])+1)
    return out
```

```python
# Parte 1

out_1=0
input=data

lista_aree = [area(input[i],input[j]) for i in range(len(input)) for j in range(i+1,len(input))]
out_1=max(lista_aree)

print(out_1)
```

```python
#Funzioni parte 2

import shapely


def vertici(X,Y):
    out=[(min(X[0],Y[0]),min(X[1],Y[1])),(max(X[0],Y[0]),min(X[1],Y[1])), (max(X[0],Y[0]),max(X[1],Y[1])),(min(X[0],Y[0]),max(X[1],Y[1]))]
    return out


def is_ok(vertice_1, vertice_2, forma):
    rettangolo=shapely.Polygon(vertici(vertice_1,vertice_2))
    if forma.contains(rettangolo):
        return True
    else:
        return False
```

```python
#Parte 2

out_2=0
input=data

shape = shapely.Polygon(input)

lista_aree_2 = [area(input[i],input[j]) for i in range(len(input)) for j in range(i+1,len(input)) if is_ok(input[i],input[j],shape)]

out_2=max(lista_aree_2)
print(out_2)
```
