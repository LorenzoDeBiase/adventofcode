```python
# Ingestion


def leggi_file_txt(nome_file):

    lista_righe = []
    try:
        with open(nome_file, 'r') as file:
            for riga in file:
                app_0=list(map(int,riga[:-1].replace(':','').split(' ')[0].split('x')))
                app=list(map(int,riga[:-1].replace(':','').split(' ')[1:]))
                lista_righe.append([app_0]+app)
                
    except FileNotFoundError:
        print(f"Errore: Il file {nome_file} non Ã¨ stato trovato.")
    except ValueError as e:
        print(f"Errore durante la conversione dei dati: {e}")
    return  lista_righe


nome_file = 'input_day12_modified.txt'
data= leggi_file_txt(nome_file)
```

```python
#Funzioni parte 1:
from math import prod

def is_ok(riga):
    if prod(riga[0])>= sum(riga[1:])*9:
        return True
    else:
        return False
```

```python
#Parte 1

out_1=0

for item in data:
    if is_ok(item):
        out_1+=1

print(out_1)
```
