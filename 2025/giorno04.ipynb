```python
# Ingestion


def leggi_file_txt(nome_file):

    lista_righe = []
    try:
        with open(nome_file, 'r') as file:
            for riga in file:
                lista_righe.append('.'+riga[:-1]+'.')


                
    except FileNotFoundError:
        print(f"Errore: Il file {nome_file} non Ã¨ stato trovato.")
    except ValueError as e:
        print(f"Errore durante la conversione dei dati: {e}")
    
    N=len(lista_righe[0])
    tmp='.'*N
    lista_righe= [tmp]+ lista_righe+ [tmp]
    lista_righe=[list(item) for item in lista_righe]
    return lista_righe  


nome_file = 'input_day4.txt'
nome_file_test= 'input_day4_test.txt'
data= leggi_file_txt(nome_file)
data_test=leggi_file_txt(nome_file_test)
```

```python
#Funzioni parte 1

def degree_of_roll(A, i, j):
    out=0
    if A[i-1][j-1]=='@':
        out+=1
    if A[i-1][j]=='@':
        out+=1
    if A[i-1][j+1]=='@':
        out+=1
    if A[i][j-1]=='@':
        out+=1
    if A[i][j+1]=='@':
        out+=1
    if A[i+1][j-1]=='@':
        out+=1
    if A[i+1][j]=='@':
        out+=1
    if A[i+1][j+1]=='@':
        out+=1
    return out
```

```python
#Parte 1

out_1=0
input=data
for i in range(1,len(input)-1):
    for j in range(1,len(input[0])-1):
        if input[i][j]=='@' and degree_of_roll(input, i ,j )<4:
            out_1+=1
print(out_1)
```

```python
#Parte 2
out_2=0
input=data
flag=True
while(flag):
    remove=[]
    for i in range(1,len(input)-1):
        for j in range(1,len(input[0])-1):
            if input[i][j]=='@' and degree_of_roll(input, i ,j )<4:
                remove.append((i,j))
    if len(remove)>0:
        out_2+=len(remove)
        for i,j in remove:
            input[i][j]='.'
    else:
        flag=False
print(out_2)
```
