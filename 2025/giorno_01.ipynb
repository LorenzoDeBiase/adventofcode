```python
# Ingestion


def leggi_file_txt(nome_file):

    lista_righe = []
    try:
        with open(nome_file, 'r') as file:
            for riga in file:
                lista_righe.append([riga[0],int(riga[1:-1])])
                
    except FileNotFoundError:
        print(f"Errore: Il file {nome_file} non Ã¨ stato trovato.")
    except ValueError as e:
        print(f"Errore durante la conversione dei dati: {e}")
    return lista_righe  


nome_file = 'input_day1.txt'
nome_file_test= 'input_day1_test.txt'
data= leggi_file_txt(nome_file)
#data_test=leggi_file_txt(nome_file_test)
```

```python
#Parte 1

out_1=0
puntatore=50

for item in data:
    if item[0]=='R':
        puntatore+=item[1]
    elif item[0]=='L':
        puntatore-=item[1]
    if puntatore%100==0:
        out_1+=1

print(out_1)
```

```python
#Parte 2

out_2=0
puntatore=50

for item in data:
    if item[0]=='R':
        for i in range(item[1]):
            puntatore+=1
            if puntatore%100==0:
                out_2+=1
    elif item[0]=='L':
        for i in range(item[1]):
            puntatore-=1
            if puntatore%100==0:
                out_2+=1

print(out_2)
```
