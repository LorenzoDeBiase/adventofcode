```python
# Ingestion


def leggi_file_txt(nome_file):

    lista_righe = []
    try:
        with open(nome_file, 'r') as file:
            for riga in file:
                lista_righe.append([int(item) for item in riga[:-1].split(',')])
                
    except FileNotFoundError:
        print(f"Errore: Il file {nome_file} non Ã¨ stato trovato.")
    except ValueError as e:
        print(f"Errore durante la conversione dei dati: {e}")
    return  lista_righe


nome_file = 'input_day8.txt'
nome_file_test= 'input_day8_test.txt'
data= leggi_file_txt(nome_file)
data_test=leggi_file_txt(nome_file_test)
```

```python
# Funzioni parte 1

import math
import networkx as nx

out_1=1

input=data
N=1000

lista_distanze = [[input[i], input[j], math.dist(input[i],input[j])] for i in range(len(input)) for j in range(i+1,len(input))]
lista_distanze.sort(key= lambda x: x[2])
lista_archi=[ [tuple(row[0]),tuple(row[1])] for row in lista_distanze[:N]]


G=nx.Graph()
G.add_nodes_from([tuple(item) for item in input])
G.add_edges_from(lista_archi)


lista_connected=sorted([len(item) for item in list(nx.connected_components(G))], reverse=True)

for i in range(3):
    out_1*=lista_connected[i]

print(out_1)
```
