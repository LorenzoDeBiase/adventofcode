```python
# Ingestion


def leggi_file_txt(nome_file):

    lista_righe = []
    try:
        with open(nome_file, 'r') as file:
            for riga in file:
                lista_righe.append(list(riga[:-1]))
                
    except FileNotFoundError:
        print(f"Errore: Il file {nome_file} non Ã¨ stato trovato.")
    except ValueError as e:
        print(f"Errore durante la conversione dei dati: {e}")
    return  lista_righe


nome_file = 'input_day7.txt'
nome_file_test= 'input_day7_test.txt'
data= leggi_file_txt(nome_file)
data_test=leggi_file_txt(nome_file_test)
```

```python
# Parte 1

out_1=0

input=data
lista_split=[]

for i in range(len(input)):
    split=[]
    for j in range(len(input[0])):
        if input[i][j]=='S':
            inizio=(i,j)
        if input[i][j]=='^':
            split.append(j)
    lista_split.append(split)

lista_beam=[inizio[1]]
for i in range(len(lista_split)):

    tmp=[]
    for beam in lista_beam:
        if beam in lista_split[i]:
            tmp.append(beam-1)
            tmp.append(beam+1)
            out_1+=1
        else:
            tmp.append(beam)

    lista_beam=list(set(tmp))
print(out_1)

```
